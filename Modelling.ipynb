```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

df10 = pd.read_csv('preprocess_merged_data_path10.csv')
df1028 = pd.read_csv('preprocess_merged_data_path1028.csv')
df28 = pd.read_csv('preprocess_merged_data_path28.csv')
```

```python
# Kolom yang ingin Anda pertahankan
columns_to_keep = ['zenoAlg.AT_5m_C', 'zenoAlg.DP_5m_C', 'zenoAlg.RH_5m' , 'windAlg.Spd_2m_kt', 'windAlg.Dir_2m', 'ALgt_1m_cd/m2',
                  'baroAlg.QNH_1m_mb', 'baroAlg.QFE_1m_mb']

# Menjatuhkan semua kolom kecuali yang ingin dipertahankan
df_clasification_10 = df10.drop(columns=df10.columns.difference(columns_to_keep))
df_clasification_1028 = df1028.drop(columns=df1028.columns.difference(columns_to_keep))
df_clasification_28 = df28.drop(columns=df28.columns.difference(columns_to_keep))

# Tampilkan hasilnya
print(df_clasification_10, df_clasification_1028, df_clasification_28)
```

```python
df_clasification_10.rename(columns={'zenoAlg.AT_5m_C' : 'AT_10', 'zenoAlg.DP_5m_C' : 'dp_10', 'zenoAlg.RH_5m' : 'rh_10', 'windAlg.Spd_2m_kt' : 'spdWind_10', 
                                    'windAlg.Dir_2m': 'dirWind_10', 'ALgt_1m_cd/m2': 'cd_10' , 'baroAlg.QNH_1m_mb' : 'QNH_10', 'baroAlg.QFE_1m_mb': 'QFE_10'}, inplace=True)
df_clasification_1028.rename(columns={'zenoAlg.AT_5m_C' : 'AT_1028', 'zenoAlg.DP_5m_C' : 'dp_1028', 'zenoAlg.RH_5m' : 'rh_1028', 'windAlg.Spd_2m_kt' : 'spdWind_1028', 
                                    'windAlg.Dir_2m': 'dirWind_1028', 'ALgt_1m_cd/m2': 'cd_1028' , 'baroAlg.QNH_1m_mb' : 'QNH_1028', 'baroAlg.QFE_1m_mb': 'QFE_1028'}, inplace=True)
df_clasification_28.rename(columns={'zenoAlg.AT_5m_C' : 'AT_28', 'zenoAlg.DP_5m_C' : 'dp_28', 'zenoAlg.RH_5m' : 'rh_28', 'windAlg.Spd_2m_kt' : 'spdWind_28', 
                                    'windAlg.Dir_2m': 'dirWind_28', 'ALgt_1m_cd/m2': 'cd_28' , 'baroAlg.QNH_1m_mb' : 'QNH_28', 'baroAlg.QFE_1m_mb': 'QFE_28'}, inplace=True)
```

```python
def evaluate_models(x, y, test_size=0.2, random_state=0):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=random_state)
    # Initialize models
    models = {
        'Linear Regression': LinearRegression(),
        # 'Ridge Regression': Ridge(),
        # 'Lasso Regression': Lasso(),
        'Random Forest': RandomForestRegressor(),
        # 'Support Vector Machine': SVR(),
        'Decision Tree': DecisionTreeRegressor(),
        'K-Nearest Neighbors': KNeighborsRegressor()
    }
    
    for model_name, model in models.items():
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

        # Evaluate the model
        r2 = r2_score(y_test, y_pred)

        # Convert the R-squared score to a percentage
        percentage_score = r2 * 100

        print(f"{model_name} - Akurasi Skor: {percentage_score:.2f}%")


```

```python
x_speedwind = df_cls[['spdWind_10', 'spdWind_28']]
y_speedwind = df_cls['spdWind_1028']

evaluate_models(x_speedwind, y_speedwind)
```

```python
x_dynamic = df_cls[['dirWind_10', 'dirWind_28']]
y_dynamic = df_cls['dirWind_1028']

evaluate_models(x_dynamic, y_dynamic)
```
