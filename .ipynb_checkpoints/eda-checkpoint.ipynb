```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from windrose import WindroseAxes
df10 = pd.read_csv('preprocess_merged_data_path10.csv')
df1028 = pd.read_csv('preprocess_merged_data_path1028.csv')
df28 = pd.read_csv('preprocess_merged_data_path28.csv')
df10.head(5)
```

```python
df10.info()
```

### Wind Rose 

#### Wind Rose Runway 10 

```python
# Create a wind rose plot
ax = WindroseAxes.from_ax()
ax.bar(df10['windAlg.Dir_2m'], df10['windAlg.Spd_2m_kt'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend()

# Set title and show the plot
plt.title('Wind Rose Plot')
plt.show()
```

#### Wind Rose Runway 1028 

```python
ax = WindroseAxes.from_ax()
ax.bar(df1028['windAlg.Dir_2m'], df1028['windAlg.Spd_2m_kt'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend()

# Set title and show the plot
plt.title('Wind Rose Plot')
plt.show()
```

#### Wind Rose Runway 28 

```python
ax = WindroseAxes.from_ax()
ax.bar(df28['windAlg.Dir_2m'], df28['windAlg.Spd_2m_kt'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend()

# Set title and show the plot
plt.title('Wind Rose Plot')
plt.show()
```

### Korelasi 

#### Runway 10

```python
columns_to_keep = ['zenoAlg.AT_5m_C', 'zenoAlg.DP_5m_C', 'zenoAlg.RH_5m' , 'windAlg.Spd_2m_kt', 'windAlg.Dir_2m', 
                  'baroAlg.QNH_1m_mb', 'baroAlg.QFE_1m_mb']
df_selected_10 = df10[columns_to_keep]
plt.figure(figsize=(12,10))
sns.heatmap(df_selected_10.corr(),cmap='YlGnBu', annot=True)
plt.show()
```
